```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
import pandas as pd


csv = pd.read_csv('./imgtrain_Abnormality_and_Location_Labels.csv')

columns = csv.columns.tolist()

dict = {}

for i in range(1, len(columns)):
    indicator = columns[i].split('*')[0]
    if(indicator in dict):
        dict[indicator] += 1
    else:
        dict[indicator] = 1

```

```python
locations = []
for i in range(1, len(columns)):
    indicator = columns[i].split('*')[0]
    locations.append(columns[i].split('*')[1])

    if(i == len(columns)-1 or columns[i + 1].split('*')[0] != indicator):
        break
```

```python
indicators = list(dict.keys())
```

```python
indicators = ["bandlike_or_linear",
"groundglass",
"honeycombing",
"reticulation",
"tree_in_bud",
"airspace_disease",
"air_trapping",
"aspiration",
"atelectasis",
"bronchial_wall_thickening",
"bronchiectasis",
"bronchiolectasis",
"bronchiolitis",
"bronchitis",
"emphysema",
"hemothorax",
"interstitial_lung_disease",
"lung_resection",
"mucous_plugging",
"pleural_effusion",
"pleural_thickening",
"pneumonia",
"pneumonitis",
"pneumothorax",
"pulmonary_edema",
"septal_thickening",
"calcification",
"cancer",
"consolidation",
"cyst",
"density",
"fibrosis",
"granuloma",
"infection",
"lesion",
"inflammation",
"mass",
"lymphadenopathy",
"nodule",
"opacity",
"scarring"]
```

```python
locations = ['left_upper', 'left_mid', 'left_lower', 'right_upper', 'right_mid', 'right_lower', 'right_lung', 'left_lung', 'lung', 'airways']
```

Number of samples for all locations(including lobes) for the specific indicators

```python
for j in range(len(indicators)):
    numbers = [0 for l in range(len(locations))]
    for i in range(len(csv)):
        for k in range(len(locations)):
            column_name = f'{indicators[j]}*{locations[k]}'
            numbers[k] += csv.loc[i,column_name]
    plt.bar(locations, numbers)
    plt.xlabel('locations')
    plt.xticks(rotation=90)
    plt.ylabel('numbers')
    plt.title(indicators[j])
    plt.show()
```

Number of samples in the left vs right lung

```python
for j in range(len(indicators)):
    numbers = [0 for l in range(2)]
    short_loc = ['right_lung', 'left_lung']
    for i in range(len(csv)):
        for k in range(len(['right_lung', 'left_lung'])):
            column_name = f'{indicators[j]}*{short_loc[k]}'
            numbers[k] += csv.loc[i,column_name]
    plt.bar(['right_lung', 'left_lung'], numbers)
    plt.xlabel('locations')
    plt.xticks(rotation=90)
    plt.ylabel('numbers')
    plt.title(indicators[j])
    plt.show()
```

Number of samples per indicator in the lung region

```python
numbers_indicators = [0 for m in range(len(indicators))]
for j in range(len(indicators)):
    total = 0
    for i in range(len(csv)):
        number = 0
        for k in range(1):
            column_name = f'{indicators[j]}*lung'
            number = max(csv.loc[i, column_name], number)
        total += number
    numbers_indicators[j] = total
    
plt.bar(indicators, numbers_indicators)
plt.xlabel('indicators')
plt.xticks(rotation=90)
plt.ylabel('numbers')
plt.title('N sample for each indicator')
plt.show()
```

Number of samples per indicator in the airways region

```python
numbers_indicators = [0 for m in range(len(indicators))]
for j in range(len(indicators)):
    total = 0
    for i in range(len(csv)):
        number = 0
        for k in range(1):
            column_name = f'{indicators[j]}*airways'
            number = max(csv.loc[i, column_name], number)
        total += number
    numbers_indicators[j] = total
    
plt.bar(indicators, numbers_indicators)
plt.xlabel('indicators')
plt.xticks(rotation=90)
plt.ylabel('numbers')
plt.title('N sample for each indicator')
plt.show()
```
