```python
import nrrd
import numpy as np
%matplotlib inline
import nibabel as nib
from skimage.measure import regionprops
```

```python
nrrd_file_path = '/content/drive/MyDrive/test_slic.nrrd'

data, header = nrrd.read(nrrd_file_path)

numpy_array = np.array(data)
```

```python
labels = np.unique(numpy_array)
labels = labels[1:]
```

```python
centers = []
```

```python
for label in labels:
    label_indices = np.where(numpy_array == label)
    center = []
    for current in label_indices:
        center.append(np.average(current))
    centers.append(center)
print(centers)
centers = np.array(centers).astype(int)
```

```python
centers = []
props = regionprops(numpy_array.astype(int))
print(len(props))
for prop in props:
  x,y,z = prop.centroid
  centers.append([int(x),int(y),int(z)])
print(centers)
centers = np.array(centers)
```

```python


labels = lungs[centers[:, 0], centers[:, 1], centers[:, 2]]

left_lung_indices = np.where((labels == 1) | (labels == 2))

left_lung = centers[left_lung_indices]

left_lung = left_lung.tolist()

print(left_lung)
```

```python

labels = lungs[centers[:, 0], centers[:, 1], centers[:, 2]]

right_lung_indices = np.where((labels == 3) | (labels == 4) | (labels == 5))

right_lung = centers[right_lung_indices]

right_lung = right_lung.tolist()

print(right_lung)
```

```python
from sklearn.neighbors import NearestNeighbors
import numpy as np

k_neighbors = 5
knn_model = NearestNeighbors(n_neighbors=k_neighbors)

knn_model.fit(left_lung)

distances, indices = knn_model.kneighbors(left_lung)
indices = indices[:, 1:]
print("Indices of k-nearest neighbors:", indices)
print("Distances to k-nearest neighbors:", distances)
```

```python
k_neighbors = 5
knn_model = NearestNeighbors(n_neighbors=k_neighbors)

knn_model.fit(right_lung)

distances, indices_right = knn_model.kneighbors(right_lung)
indices_right = indices_right[:, 1:]
print("Indices of k-nearest neighbors:", indices)
print("Distances to k-nearest neighbors:", distances)
```

```python
edges = []
for i in range(len(indices)):
    for column in indices[i]:
        edges.append([left_lung[i], left_lung[column]])
for i in range(len(indices_right)):
    for column in indices_right[i]:
        edges.append([right_lung[i], right_lung[column]])
```
